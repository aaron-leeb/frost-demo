# FROST Protocol: Participants & Variables Overview

This notebook provides a high-level overview of the FROST threshold signature protocol, focusing on the roles of participants and the key variables at each stage.

## Protocol Stages

1. **Key Generation**
2. **Preprocessing**
3. **Signing**

## Participants

- **Participant ($P_i$):** Holds a secret share $s_i$, generates nonces, and computes signature shares.
- **Signature Aggregator (SA):** Coordinates signing, collects shares, verifies, and publishes the final signature. May be one of the $P_i$.

## Variables at Each Stage

| Stage           | Participant ($P_i$) Variables         | Aggregator (SA) Variables         | Shared Variables         |
|-----------------|--------------------------------------|-----------------------------------|-------------------------|
| Key Generation  | $s_i$, $Y_i$                         | $Y$ (group key)                   | $Y$, $Y_i$              |
| Preprocessing   | $(d_i, e_i)$, $(D_i, E_i)$           | Commitment lists $L_i$            | $(D_i, E_i)$            |
| Signing         | $
ho_i$, $R_i$, $z_i$, $ambda_i$  | $B$, $R$, $c$, $z$, $igma$      | $m$, $B$, $R$, $c$, $igma$ |

## Protocol Flow Diagram

Below is a simplified depiction of the protocol flow.

```mermaid
sequenceDiagram
    participant P1
    participant P2
    participant ...
    participant SA
    Note over P1,P2,...: KeyGen & Preprocessing
    SA->>P1: Request (m, B)
    SA->>P2: Request (m, B)
    ...
    P1-->>SA: $z_1$
    P2-->>SA: $z_2$
    ...
    SA->>SA: Aggregate, Verify, Publish $(R, z)$
```

## Step-by-Step Variable Ownership

1. **Key Generation:**
    - Each $P_i$ gets $s_i$, $Y_i$; SA computes $Y$.
2. **Preprocessing:**
    - Each $P_i$ generates $(d_i, e_i)$, computes $(D_i, E_i)$, shares $(D_i, E_i)$ with SA.
3. **Signing:**
    - SA selects $S$, builds $B$, sends $(m, B)$ to each $P_i$.
    - Each $P_i$ computes $
ho_i$, $R_i$, $z_i$, and sends $z_i$ to SA.
    - SA verifies, aggregates $z = um z_i$, publishes $(R, z)$.
